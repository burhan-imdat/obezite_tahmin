```python
from pycaret.classification import*
# gerekli kütüphane indirildi
```

```python
df=pd.read_csv('train.csv')
xt=pd.read_csv('test.csv')
# dosya okundu
```

```python
df.head()
#  eğitilecek dosya içeriği görüldü
```

```python
xt.head()
# tahmin edilecek dosya içeriği görüldü
```

```python
x=df.drop('NObeyesdad',axis=1)
y=df['NObeyesdad']
# x ve y içeriği belirlendi
```

```python
idt=xt['id']
xt=xt.drop('id',axis=1)
# id yi kayıt ettik
```

```python
x=x.drop('id',axis=1)
# kullanılmayan id silindi
```

```python
xd=pd.get_dummies(x,drop_first=True)
xtd=pd.get_dummies(xt,drop_first=True)
# katodorik olanları sayısala çevrildi
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
# gerekli dosyalar indirildi
```

```python
X_train, X_test, y_train, y_test = train_test_split(xd, y, test_size=0.2, random_state=42)
# x ve y train test ayrımı yapıldı
```

```python
gb_classifier = GradientBoostingClassifier()
# iğitim nesnesi oluşturuldu
```

```python
gb_classifier.fit(X_train, y_train)
```

```python
X_test.head()
# içerik görüldü
```

```python
xtd.head()
# içerik görüldü
```

```python
xtd.drop('CALC_Frequently',axis=1,inplace=True)
# fazla olan alan düşürüldü
```

```python
predictions = gb_classifier.predict(X_test)
# model değerlendirmek için tahmin yapıldı
```

```python
print(classification_report(y_test, predictions))
#model değerlendirmek için tahmin yapıldı
```

```python
tahmin=pd.DataFrame({'id':idt,'NObeyesdad':gb_classifier.predict(xtd)}).set_index('id')
# tahmin kaydı için sözlük yapısı oluşturuldu
```

```python
tahmin.to_csv('submission.csv')
# taahmin kayıt edildi
```
